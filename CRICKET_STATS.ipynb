```python
matches["Win_Type"] = np.where(
    matches["win_by_runs"]>0,
    "Batting",
    "Bowling"
)
```

```python
deliveries["total_ball_runs"] = (
    deliveries["batsman_runs"] +
    deliveries["extra_runs"]
)
```

```python
ipl = deliveries.merge(
    matches,
    left_on="match_id",
    right_on="id"
)
```

```python
plt.figure(figsize=(10,5))#matches per season by countplot
sns.countplot(
    data=matches,
    x="Season"
)
plt.xticks(rotation=45)
plt.show()

matches["winner"].value_counts()#number of wins by every team in all seasons

```

```python
plt.figure(figsize=(12,6))#total win by teams in all seasons in countplot
sns.countplot(
    y="winner",
    data=matches,
    order=matches["winner"].value_counts().index
)
plt.show()

matches.groupby("venue")["winner"].count().sort_values(ascending=False)# total winnings in every stadium
```

```python
top = deliveries.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10)
top.plot(kind="bar")#runs in ipl by batsmans in bar graph
```

```python
deliveries.groupby(
["match_id","batsman"]
)["batsman_runs"].sum().sort_values(ascending=False).head(10)#highest runs batted in single inning by a player
```

```python
runs = deliveries.merge(matches,left_on="match_id",right_on="id")#orange cups in every season

orange = runs.groupby(
["Season","batsman"]
)["batsman_runs"].sum()

orange.groupby(level=0).nlargest(1)
```

```python
deliveries[
deliveries["player_dismissed"].notna()
]["bowler"].value_counts().head(10)#total wickets by bowlers in every season top 10
```

```python
#economy rank by bowlers
bowler = deliveries.groupby("bowler").agg(
runs=("total_runs","sum"),
balls=("ball","count")
)
bowler["overs"]=bowler["balls"]/6
bowler["economy"]=bowler["runs"]/bowler["overs"]
bowler.sort_values("economy").head(20)
```

```python
#highest strike rate by players
bat = deliveries.groupby("batsman").agg(
runs=("batsman_runs","sum"),
balls=("ball","count")
)

bat["SR"]=bat["runs"]/bat["balls"]*100

bat.sort_values("SR",ascending=False)
```

```python
#highest runs by team in single match
team_score = deliveries.groupby(
["match_id","inning","batting_team"]
)["total_runs"].sum()

team_score.sort_values(ascending=False).head(10)
```

```python
#lowest runs by a team in single match
team_score.sort_values().head(10)
```

```python
#win percentage by every teams
played = pd.concat([
matches["team1"],
matches["team2"]
]).value_counts()
won = matches["winner"].value_counts()
win_percent = (won/played)*100
win_percent.sort_values()
```
